```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib as mpl
from scipy.stats import chi2_contingency
import matplotlib as mpl
import matplotlib.pyplot as plt      
import math
from collections import Counter
from sklearn.decomposition import PCA
%matplotlib inline
from matplotlib import cm
```

```python
df = pd.read_csv('data cleaning new1.csv',index_col=0) 
df
```

```python
plt.gcf().set_size_inches((10, 10))    
df["Numbers_for_Type_house_in_df"].value_counts().plot(kind='bar')
plt.title('Number of Apartments by type house')
```

```python
ax = plt.gca()
df.plot(kind='scatter',x='Floor',y='Price', color='red', ax=ax)
plt.title('The connection between num of floor and price')
```

```python
plt.scatter(df["Num of rooms"], df["Price"])
plt.xlabel('Num of rooms')
plt.ylabel('Price')
plt.title('The connection between num of rooms and price')
plt.axis([0,8.5, 250000, 4000000])
plt.gcf().set_size_inches((10, 10))   
plt.show()
```

```python
plt.scatter(df["Size of house per mm"], df["Price"],color='orange')
plt.axis([0, 600, 250000, 8000000])
plt.gcf().set_size_inches((10, 10))  
plt.title('The connection between Size of house and price')
plt.xlabel("Size of house per mm")
plt.ylabel("Price")
plt.show()
```

```python
plt.scatter(x=df["Numbers_for_Type_house_in_df"], y=df["Price"],color='green',marker='*')
plt.xlabel('Numbers_for_Type_house_in_df')
plt.title('The connection between Size of house and price')
plt.ylabel('Price')
plt.axis([-1, 20, 250000, 8000000])
plt.show()
```

```python
plt.scatter(x=df["Numbers_for_Cities_in_df"], y=df["Price"],c='c',marker='*')
plt.xlabel('Numbers_for_Cities_in_df')
plt.ylabel('Price')
plt.gcf().set_size_inches((10, 10))  
plt.title('The connection between the city and price')
plt.show()
```

```python
df["Numbers_for_Cities_in_df"].value_counts().plot(kind='bar' ,figsize=(350, 20))
plt.title('Number of Apartments by City')
```

```python
df["Numbers_for_Cities_in_df"].value_counts().plot(kind='pie' ,figsize=(15, 15))
plt.title('Number of Apartments by City')
```
