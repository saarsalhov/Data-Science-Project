```python
import time      # for testing use only
import os         # for testing use only

import bs4
import requests
from bs4 import BeautifulSoup  
import pandas as pd
import scipy as sc
import numpy as np
```

```python
homepage = 'https://www.ad.co.il'
page1= requests.get(homepage + '/nadlansale/')
soup = BeautifulSoup(page1.text, 'html.parser')
```

```python
def check_if_exist(lenth):
    if (lenth == 3):
         exist = 1 #not exist
    else:
        exist = 0 #exist
    return exist
```

```python
type_house=[] #סוג הנכס
city=[] # עיר
street=[] #רחוב
prices=[] #מחיר
num_of_rooms=[] #מספר חדרים
floors=[] #קומה
elevator=[] #מעלית
furniture=[] #רהיטים
bars=[] #סורגים
terrace=[] #מרפסת
parking=[] #יש חניה או אין
size_per_mm =[] #גודל הדירה במ"מ
n_pages =1869
main_data =soup.find_all('div',attrs={'class':'card overflow-hidden'})
numberPage = 1
    for lst in main_data:
        all_apartment_pages= lst.find('a')
        html_doc = homepage + all_apartment_pages.get("href")
        apartment_page = requests.get(html_doc)
        details_soup = BeautifulSoup(apartment_page.text, "html.parser")
        details_Block = details_soup.find('div',attrs={'class':'single-product-tab tab-pane fade show active'})  
        price = details_Block.find_all('h2', attrs={'class':'card-title'})
        floor = details_Block.find_all('span', attrs={'class':'ms-1'})
        details_Block2=details_soup.find('table',attrs={'class':'table table-sm mb-4'})
        tabel_details = details_Block2.find_all('td')
        last_block = details_soup.find('div',attrs={'class':'card-icons flex-wrap d-flex h-100'})
        last_block_details = last_block.find_all('div')
        print(tabel_details[5].get_text())
        print(tabel_details[9].get_text())
        if(len(price) == 2 and len(floor) == 3):
            prices.append(price[1].get_text())
            floors.append(floor[2].get_text())
            num_of_rooms.append(floor[0].get_text())
            size_per_mm.append(floor[1].get_text())
            type_house.append(tabel_details[1].get_text())
            city.append(tabel_details[5].get_text())
            street.append(tabel_details[9].get_text())
            furniture.append(check_if_exist((len(last_block_details[1].get('class')))))
            parking.append(check_if_exist(len(last_block_details[3].get('class'))))
            terrace.append(check_if_exist(len(last_block_details[5].get('class'))))
            bars.append(check_if_exist(len(last_block_details[7].get('class'))))
            elevator.append(check_if_exist(len(last_block_details[8].get('class'))))
    end_block = soup.find('li', attrs={'class':'page-item mx-1 nextPage'})
    next_p = end_block.find('a', attrs={'class':'page-link'} )
    html_dock_new = homepage + next_p.get("href")
    n_page= requests.get(html_dock_new)
    soup = BeautifulSoup(n_page.text, "html.parser")
    numberPage = numberPage + 1
    main_data = soup.find_all('div',attrs={'class':'card overflow-hidden'})
df=pd.DataFrame({'Type house': type_house ,  'City': city, 'Street':street, 'Num of rooms': num_of_rooms ,  'Floor': floors,
                 'Price':prices,  'Bars': bars,  'Furniture': furniture,  'Elevator': elevator,  'Terrace': terrace
                ,  'Parking': parking,  'Size of house per mm': size_per_mm})
df

```
